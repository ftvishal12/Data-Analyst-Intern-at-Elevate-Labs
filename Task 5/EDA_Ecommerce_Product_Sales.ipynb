```python

# Exploratory Data Analysis on Ecommerce Product Sales

## 1. Dataset Overview
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/mnt/data/Ecommerce_Product_Sales.csv")
df.info()
df.describe()
df.head()

## 2. Missing Values Check
df.isnull().sum()

## 3. Univariate Analysis

### a. Distribution of Price
sns.histplot(df['Price'], kde=True)
plt.title("Distribution of Price")
plt.show()

### b. Units Sold Distribution
sns.histplot(df['Units_Sold'], kde=True)
plt.title("Distribution of Units Sold")
plt.show()

### c. Ratings Distribution
sns.histplot(df['Rating'], kde=True)
plt.title("Distribution of Ratings")
plt.show()

### d. Return Rate Distribution
sns.histplot(df['Return_Rate'], kde=True)
plt.title("Distribution of Return Rate")
plt.show()


### e. Category Count
df['Category'].value_counts().plot(kind='bar')
plt.title("Count of Products by Category")
plt.xticks(rotation=45)
plt.show()

### f. Seller Count
df['Seller_Name'].value_counts().plot(kind='bar', color='orange')
plt.title("Count of Products by Seller")
plt.xticks(rotation=45)
plt.show()


## 4. Bivariate Analysis

### a. Correlation Heatmap
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()


### b. Pairplot of Numeric Features
sns.pairplot(df[['Price', 'Units_Sold', 'Revenue', 'Rating', 'Return_Rate']])
plt.show()

### c. Scatter Plot: Revenue vs Units Sold
sns.scatterplot(data=df, x='Units_Sold', y='Revenue')
plt.title("Revenue vs Units Sold")
plt.show()


### d. Boxplot: Price by Category
sns.boxplot(data=df, x='Category', y='Price')
plt.title("Price by Category")
plt.xticks(rotation=45)
plt.show()




```
